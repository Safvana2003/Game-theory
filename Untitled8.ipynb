```python
# Simple Game Theory Example: Prisoner's Dilemma

payoff_matrix = {
    ('C', 'C'): (3, 3),
    ('C', 'D'): (0, 5),
    ('D', 'C'): (5, 0),
    ('D', 'D'): (1, 1)
}

# All possible strategies
strategies = ['C', 'D']

# Function to find Nash Equilibrium
def find_nash_equilibrium():
    equilibria = []
    for a_strategy in strategies:
        for b_strategy in strategies:
            a_payoff = payoff_matrix[(a_strategy, b_strategy)][0]
            b_payoff = payoff_matrix[(a_strategy, b_strategy)][1]


            a_best_response = True
            for alt_a in strategies:
                if payoff_matrix[(alt_a, b_strategy)][0] > a_payoff:
                    a_best_response = False

            b_best_response = True
            for alt_b in strategies:
                if payoff_matrix[(a_strategy, alt_b)][1] > b_payoff:
                    b_best_response = False

            if a_best_response and b_best_response:
                equilibria.append((a_strategy, b_strategy))

    return equilibria


nash_eq = find_nash_equilibrium()

print("Prisoner's Dilemma - Payoff Matrix:")
for move, payoff in payoff_matrix.items():
    print(f"{move}: {payoff}")

print("\nNash Equilibrium:", nash_eq)
```
